```python
def create_data(year):
    datas = [None, None, None]  # 빈 리스트로 초기화

    for file_num in range(1, 4, 1):
        df = pd.read_excel("../Data/순위.xlsx", sheet_name=f"{year}_{file_num}")
        for i in range(0, df.shape[0], 2):
            df.iloc[i, 1] = df.iloc[i + 1, 1]
        delete_rows = [i for i in range(1, df.shape[0], 2)]
        df.drop(delete_rows, inplace=True)
        if file_num in [2, 3]:
            df.drop(axis=1, columns=["순위"], inplace=True)
        datas[file_num - 1] = df.reset_index(drop=True)

        if file_num == 2:
            merged_df = pd.merge(datas[0], datas[1], on="팀", how="inner")
            
        if file_num == 3:
            result = pd.merge(merged_df, datas[2], on="팀", how="inner")
            result = result.rename(columns=lambda x: x.replace('정렬', '') if '정렬' in x else x)
            result["연도"]=year
            return result
```

```python
data_07_23=[]
for i in range(2007,2024):
    print(i)
    data_07_23.append(create_data(i))
df=pd.concat([data_07_23[i] for i in range(17)], axis=0, ignore_index=True)
```

```python
df.to_csv("baseballTeam.csv", index=False)
```
