```python
import matplotlib.pyplot as plt
import numpy as np
import koreanize_matplotlib

# 순위는 : 중간 정도의 선형 관계 WHIP, (평균자책,실점,자책) , 세이브와 연관이 있다
%matplotlib inline
```

```python
import pymysql
import pandas as pd


# 파이썬 파일에서 Mysql 과 연결
conn=pymysql.connect(host='localhost', user='root', password='1234', db='testdb', charset='utf8')

cur = conn.cursor() # cursor생성 : SQL문 실행, 결과 집합 처리, 트랜잭션 관리, 오류 처리를 수행함
cur.execute('select * from baseballTeam') # SQL문 실행
rows=cur.fetchall() # 결과 집합 처리
df=pd.DataFrame(rows) # DataFrame으로 변환
```

```python
df
```

```python
# 순위는 : 중간 정도의 선형 관계 WHIP, (평균자책,실점,자책) , 세이브와 연관이 있다
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10),
                         sharex=True, sharey=True)
ax=axes[0]
a = np.array(df[0])
b = np.array(df[33])

ax = fig.add_subplot(111)
ax.scatter(a,b)


poly_fit=np.polyfit(a,b,1)
poly_1d=np.poly1d(poly_fit)
y=poly_1d(a)
ax.plot(a,y,color='gray')

ax.set_xlabel('순위')
ax.set_ylabel('WHIP(투수가 얼마나 많은 주자들에게 베이스를 허용하는지)')
ax.set_title("WHIP과 순위의 산점도")
ax.set_xticks([])  # x축 눈금 없애기
ax.set_yticks([])  # y축 눈금 없애기
plt.show()
```
