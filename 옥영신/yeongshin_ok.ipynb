```python
a = [df['홈런'], df['타수'], df['타율'], df['투구수/타석']]
b = df['출루율+장타율']
c = ['홈런', '타수', '타율', '투구수/타석']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
axes = axes.flatten()  # 2차원 axes 배열을 1차원 배열로 변환

for i in range(len(a)):
    poly_fit = np.polyfit(a[i], b, 1)
    poly_1d = np.poly1d(poly_fit)
    xs = np.linspace(a[i].min(), a[i].max(), 100)  # xs를 a[i]의 범위에 맞게 생성
    ys = poly_1d(xs)
    
    ax = axes[i]  # 현재 서브플롯 선택
    ax.scatter(a[i], b, color='yellow')
    ax.plot(xs, ys, color='red')
    ax.set_title(f'{c[i]} & OPS 간의 상관관계', size=25)
    ax.set_xlabel(f'{c[i]}', size=20)
    ax.set_ylabel('OPS', size=20)
    ax.tick_params(axis='x', labelsize=15)
    ax.tick_params(axis='y', labelsize=15)
    ax.text(1,1,s='correlation: ' + str(round(np.corrcoef(a[i], b)[0,1], 2)))
    
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import platform
from matplotlib import font_manager, rc

# 글자 깨지는 부분 방지용도
if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:
    rc('font', family='AppleGothic')

fig = plt.figure(figsize=(20, 15))

variables = ['홈런', '타수', '출루율+장타율', '투구수/타석']
titles = ['홈런 별 분포', '타수 별 분포', 'OPS 별 분포', '투구수/타석 별 분포']

for i, var in enumerate(variables):
    plt.subplot(2, 2, i+1)
    # 데이터의 히스토그램
    data = df[var].dropna()  # NA 값 제거
    plt.hist(data, bins=20, alpha=0.6, color='g')
    
    # 정규 분포 곡선을 위한 데이터 생성
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, data.mean(), data.std())
    
    # 정규 분포 곡선 추가
    plt.plot(x, p, linewidth=2, color='red')
    title = titles[i]
    plt.title(title, size=25)
    plt.xlabel(var, size=20)
    plt.ylabel('선수 수', size=20)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)

plt.tight_layout()
plt.show()
```

```python
final_data=pd.merge(pd.DataFrame(toa),td, on='팀명',how="inner")
final_data.rename(columns={0:'평균OPS'}, inplace=True)
final_data
```

```python
import numpy as np
import matplotlib.pyplot as plt
if platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
    rc('font', family=font_name)
else:
    rc('font', family='AppleGothic')
# 가정: 'toa'와 'td'는 pandas DataFrame, 이미 정의되어 있으며 '승률'이라는 컬럼을 포함하고 있음
# 가정: 'c' 리스트는 그래프 제목과 X축 라벨을 위한 문자열 리스트

# 단일 서브플롯 생성
fig, ax = plt.subplots(figsize=(15, 15))

# toa와 td['승률']에서 .values 속성을 사용하여 numpy 배열로 변환
x_values = final_data['평균OPS'].values
y_values = final_data['승률'].values

# 2차 회귀곡선으로 변경
poly_fit = np.polyfit(x_values, y_values, 2)
poly_1d = np.poly1d(poly_fit)
xs = np.linspace(x_values.min(), x_values.max(), 100)
ys = poly_1d(xs)

# 회귀곡선 그리기
ax.scatter(x_values[:], y_values[:], color='blue')
ax.plot(xs, ys, color='red')
ax.set_title('팀 평균OPS & 팀 승률 간의 상관관계', size=25)
ax.set_xlabel('팀 평균OPS', size=20)
ax.set_ylabel('팀 승률', size=20)
ax.tick_params(axis='both', which='major', labelsize=15)

# 상관계수 계산 및 텍스트 표시
correlation = np.corrcoef(x_values, y_values)[0, 1]
ax.text(xs.min(), ys.max(), f'correlation: {round(correlation, 2)}', fontsize=12)

# 서브플롯 간격 조정 및 레이아웃 맞춤
plt.tight_layout()
plt.show()
```
